```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from tensorflow.keras.datasets import mnist
```

## Step 1: Load Dataset

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print("Training data shape:", X_train.shape)
print("Test data shape:", X_test.shape)
```

## Step 2: Flatten Images

```python
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

print("Flattened training shape:", X_train_flat.shape)
```

## Step 3: Apply PCA

```python
pca = PCA(n_components=100)
pca.fit(X_train_flat)
```

## Step 4: Reconstruction

```python
X_train_pca = pca.transform(X_train_flat)
X_train_reconstructed = pca.inverse_transform(X_train_pca)
```

## Step 5: Visualization

```python
n_images = 5
plt.figure(figsize=(10, 4))

for i in range(n_images):
    # Original
    plt.subplot(2, n_images, i + 1)
    plt.imshow(X_train[i], cmap="gray")
    plt.title("Original")
    plt.axis("off")

    # Reconstructed
    plt.subplot(2, n_images, i + 1 + n_images)
    plt.imshow(X_train_reconstructed[i].reshape(28, 28), cmap="gray")
    plt.title("Reconstructed")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

## Reflection
- What information is lost during PCA?
Fine pixel-level details
Stroke sharpness
Noise and small variations


- Why does reconstruction blur fine details?
PCA keeps only top variance directions
High-frequency details are discarded
Inverse transform is an approximation
